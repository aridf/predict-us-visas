```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno

%matplotlib inline
plt.rcParams['axes.linewidth'] = 2
```

```python
df = pd.read_csv("data/us_perm_visas.csv", parse_dates=True)
shape = df.shape
print(shape)
```

```python
df.case_status.value_counts(dropna=False)
```

```python
pct_na = (df.drop('case_status', 1).isna().sum() / shape[0]) * 100

counts, bins, patches = plt.hist(pct_na, 10, alpha = 0.75)

plt.xticks(np.arange(0, 101, 10))
plt.title('Distribution of % NAn among variables')
plt.xlabel('% NAn')
plt.ylabel('# variables')
plt.show()
```

```python
msno.matrix(df.sample(100))

plt.title('Matrix of missing observations')
plt.xlabel('Variables')
plt.ylabel('Observations')
plt.show()
```

```python
pct_na[pct_na < 11]
```

```python
df.class_of_admission.value_counts(dropna=False)
```

```python
h1_b = df['class_of_admission']=='H-1B'
df_red = df[h1_b]
df_red.class_of_admission.value_counts(dropna=False)
```

```python
msno.matrix(df_red.sample(100))

plt.title('Matrix of missing observations')
plt.xlabel('Variables')
plt.ylabel('Observations')
plt.show()
```

```python
df_red_sample = df_red.sample(50)
df_red_sample['null_count'] = df_red_sample.isnull().sum(axis=1)
df_red_sample = df_red_sample.sort_values(by = 'null_count')

msno.matrix(df_red_sample.loc[ : , 0:15])

plt.title('Matrix of missing observations')
plt.xlabel('Variables')
plt.ylabel('Observations')
plt.show()
```
