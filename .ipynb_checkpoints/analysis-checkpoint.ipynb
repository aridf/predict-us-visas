## xgboost preprocessing

```python
# trim features for quick and dirty model
df = df.loc[:,["employer_state", "applicant_major", "applicant_education", "citizenship", 'wage', 'case_status']]
```

```python
y = df['case_status']
df = df.drop('case_status', axis = 1)
```

```python
# encode y as integer
label_encoder = LabelEncoder()
label_encoder = label_encoder.fit(y)
y = label_encoder.transform(y)
print(y)
```

```python
X = df.select_dtypes('float').values
```

```python
X_to_encode = df.select_dtypes(exclude = ['float']).values

for i in range(0, X_to_encode.shape[1]):
    label_encoder = LabelEncoder()
    feature = label_encoder.fit_transform(X[:,i])
    feature = feature.reshape(X.shape[0], 1)
    onehot_encoder = OneHotEncoder(sparse=False, categories='auto')
    feature = onehot_encoder.fit_transform(feature)
    X = np.concatenate((X, feature), axis=1)
```

```python
seed = 10
test_size = 0.25

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size)

D_train = xgb.DMatrix(X_train, label = y_train)
D_test = xgb.DMatrix(X_test, label = y_test)
```

```python
# fit model to training data
params = {
    'eta': 0.3,
    'max_depth': 3,
    'objective': 'multi:softprob',
    'num_class': 3
}

steps = 20

model = xgb.train(params, D_train, steps)
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

preds = model.predict(D_test)
best_preds = np.asarray([np.argmax(line) for line in preds])

#get precision, recall, accuracy of the model from test
#weighted average accounts for nonuniformity of feature distributions 
print("Precision = {}".format(precision_score(y_test, best_preds, average='weighted')))
print("Recall = {}".format(recall_score(y_test, best_preds, average='weighted')))
print("Accuracy = {}".format(accuracy_score(y_test, best_preds)))
```
